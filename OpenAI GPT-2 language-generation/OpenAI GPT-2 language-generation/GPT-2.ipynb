```python
import torch
from transformers import GPT2Tokenizer, GPT2LMHeadModel
```

```python
tokenizer= GPT2Tokenizer.from_pretrained('gpt2')
model = GPT2LMHeadModel.from_pretrained('gpt2')
```

```python
sequence=("Language generation is one of those natural language tasks that can really produce an incredible feeling of awe at how far the fields of machine learning and artificial intelligence have come.")
```

```python
sequence
```

```python
inputs = tokenizer.encode(sequence, return_tensors='pt')
```

```python
outputs= model.generate(inputs, max_length=200, do_sample=True )
```

```python
text= tokenizer.decode(outputs[0], skip_special_tokens=True)
text
```

```python
outputs= model.generate(inputs, max_length=200, do_sample=True, temperature=.7, top_k=50 )
```

```python
text= tokenizer.decode(outputs[0], skip_special_tokens=True)
text
```
